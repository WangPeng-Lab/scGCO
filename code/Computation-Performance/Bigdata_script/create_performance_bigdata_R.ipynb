```r
library('trendsceek')
```

```r
#using trendsceek sim
set.seed(12346)
pp=sim_pois(1000000)  ## number of cells

low_expr=sample(1:10,100,replace=T)  # number of genes
high_expr=sample(20:50,100,replace=T)

pp=add_markdist_step(pp,low_expr,high_expr)
```

```r
data=pp$marks[1:1000000,]
```

```r
## combine corrd (x 'x' y)
X=pp$x
Y=pp$y
info=paste0(X,'x',Y)
info=info[1:1000000]
```

```r
row.names(data)=info
```
